Funciones en Python

Funciones básicas

```python
def saludar():
    print("Hola")

saludar()
```

Funciones con argumentos

```python
def mostrar_nombre(nombre):
    print("Mi nombre es", nombre)

mostrar_nombre("María")
mostrar_nombre("Enrique")
mostrar_nombre("Nando")
```

Funciones con argumentos por posición

```python
def resta(a, b):
    print(a - b)

resta(3,5) #-2
resta(5,3) #2

def suma(a,b):
    print(a+b)

suma(3,5) #8
suma(5,3) #8
    
```

Funciones con argumentos por nombre

```python
def multiplicacion(a,b):
    print(a*b)

multiplicacion(b = 5, a = 8) #40

#Retomando la función de resta previamente creada
resta(b = 5, a = 7) #2
resta(7, 5) #2
```

```python
def presentar(nombre, edad):
    print("Nombre: ", nombre)
    print("Edad: ", edad)

presentar(nombre="Julián", edad=50)
presentar(edad=20, nombre="Sofía")
```

Funciones con argumentos por defecto

```python
def suma(a = 5, b = 4, c = 23):
    return a + b + c

print(suma())
print(suma(1, 2, 3))
```

```python
def presentar(nombre="José", ciudad="Veracruz"):
    return "Nombre: " + nombre + ", Ciudad: " + ciudad

print(presentar())
print(presentar("Juan", "Monterrey"))
print(presentar(nombre="Juan", ciudad="Naucalpan"))
```

Funciones con argumentos de longitud variable

```python
def sumar(lista_numeros):
    total = 0
    for n in lista_numeros:
        total += n
    return total

print(sumar([5, 4, 6]))
numeros_db = [5, 5, 5, 5, 5, 5]
print(sumar(numeros_db))
numeros_db = [10, 1]
print(sumar(numeros_db))
```

```python
def promedio(lista_numeros):
    total = 0
    for n in lista_numeros:
        total += n
    return total / len(lista_numeros)

calificaciones_db = [6.7, 9.8, 8.8, 10, 8.9, 8.9]
print("Promedio: " + str(promedio(calificaciones_db)))
```

```python
def contar_palabras(lista_textos):
    total = 0
    for palabra in lista_textos:
        total += 1
    return total

print(contar_palabras(["Hola", "Mundo", "en", "Python"]))

def contar_palabras_dinamico(texto):
    total = 0
    palabras = texto.split() #Separar string en un array
    for i in palabras:
        total += 1
    return total

print(contar_palabras_dinamico("Hola Mundo en Python"))
print(contar_palabras_dinamico("Hola Mundo"))
```

Ejemplo Integrador

```python
#Función simple
def bienvenida():
    print("Bienvenidos al sistema")
bienvenida

#Función con argumentos
def saludar(nombre):
    print("Hola", nombre)
saludar("Pablo")

#Funcion con argumentos por posición
def resta(a, b):
    return a-b

resultado1 = resta(100, 3)
print("Resta por argumentos por posición es igual a: ", resultado1)

#Función con argumentos por nombre
resultado2 = resta(b = 15, a = 12)
print("Resta por argumentos con nombre ", resultado2)

#Función con argumentos por defecto
def calcular_total(precio, impuesto = 0.16):
    return precio + (precio*impuesto)
print("Total con impuesto por defecto: ", calcular_total(100))
print("Total con impuesto personalizado: ", calcular_total(100, 0.06))

#Funciones con argumentos de longitud variable
def sumar_numeros(numeros):
    total = 0
    for n in numeros:
        total += n
    return total
print("Suma de varios números: ", sumar_numeros([7,4,3,6,7,20]))
```
