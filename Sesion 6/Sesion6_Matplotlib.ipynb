```python
# Importar bibliotecas
import matplotlib.pyplot as plt
import numpy as np
# Graficar datos
plt.plot([-1, -4.5, 16, 23])
plt.show()
```

Al agregar una cadena de formato a la función *plot*, podemos crear un gráfico con valores discretos, en nuestro caso, marcadores circulares azules. La cadena de formato define cómo deben representarse los puntos discretos.

```python
#https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html
plt.plot([-1, -4.5, 16, 23], "ob") #o = circle marker b = blue
plt.show()
```

Con muy poca información, pudimos crear gráficos en los ejemplos anteriores. Sin embargo, faltaba información en los gráficos. Normalmente, nos gustaría asignar un nombre a los valores de "x" y "y". Además, todo el gráfico debería tener un título. Lo demostramos en el siguiente ejemplo:

```python
dias = range(1, 9)
valores_celsius = [25.6, 24.1, 26.7, 28.3, 27.5, 30.5, 32.8, 33.1]

fig, ax = plt.subplots() # Se crea una figura con un eje.
ax.plot(dias, valores_celsius)
ax.set(xlabel='Día',
       ylabel='Temperatura en Celsius',
       title='Gráfico de Temperatura')
plt.show()
fig.savefig("Figura 1.png")
```

**Verificar y definir los rangos de los ejes**

También se pueden verificar y definir los rangos de los ejes con la función *axis*. Si se invoca sin argumentos, regresa los límites actuales de los ejes.

```python
dias = range(1, 9)
valores_celsius = [25.6, 24.1, 26.7, 28.3, 27.5, 30.5, 32.8, 33.1]

fig, ax = plt.subplots() # Se crea una figura con un eje.
ax.plot(dias, valores_celsius)
ax.set(xlabel='Día',
       ylabel='Temperatura en Celsius',
       title='Gráfico de Temperatura')
print("Los límites actuales para los ejes son:")
print(ax.axis())
print("Se establecen los ejes en los siguientes valores:")
xmin, xmax, ymin, ymax = 0, 10, 14, 45
print(xmin, xmax, ymin, ymax)
ax.axis([xmin, xmax, ymin, ymax])
plt.show()
```

**Varios gráficos en uno**

Vimos que la función *plot* es necesaria para representar gráficamente una figura, o mejor dicho, las figuras, ya que puede haber más de una. Podemos especificar un número arbitrario de grupos *x*, *y*, en una función *plot*. Ampliaremos nuestro ejemplo anterior de temperatura para demostrarlo. Proporcionamos dos listas con valores de temperatura: una para los valores mínimos y otra para los máximos:

```python
dias = list(range(1,9))
celsius_min = [19.6, 24.1, 26.7, 28.3, 27.5, 30.5, 32.8, 33.1]
celsius_max = [24.8, 28.9, 31.3, 33.0, 34.9, 35.6, 38.4, 39.2]
fig, ax = plt.subplots()
ax.set(xlabel='Día',
ylabel='Temperatura en Celsius',
title='Gráfico de Temperatura')
ax.plot(dias, celsius_min,
dias, celsius_min, "oy", #o = circle marker y = yellow
dias, celsius_max,
dias, celsius_max, "or") #o = circle marker r = red
plt.show()
```

**Histogramas**

Un histograma es un tipo de gráfico de barras que ofrece una representación discretizada de la frecuencia de valores. Los puntos de datos se dividen en intervalos discretos con una distribución uniforme, y se representa gráficamente el número de puntos de datos en cada intervalo.

```python
datos = np.random.normal(size=20000) # Genera un arreglo 1-D de 10000 números
plt.hist(datos, bins=200)
plt.title("Histograma")
plt.xlabel("Valores")
plt.ylabel("Frecuencia")
plt.show()
```

**Diagramas de dispersión**

Los diagramas de dispersión son una forma útil de examinar la relación entre dos o más variables.

```python
x = np.arange(0, 12)
y1 = np.random.randint(2, 7, (12,))
y2 = np.random.randint(9, 14, (12,))
y3 = np.random.randint(15, 25, (12,))
# Markers: https://matplotlib.org/api/markers_api.html
plt.scatter(x, y1, label="y1")
plt.scatter(x, y2, marker='v', color='r', label="y2")
plt.scatter(x, y3, marker='^', color='m', label="y3")
plt.legend()
plt.title('Ejemplo de Diagrama de Dispersión')
plt.show()
```

**Gráficos apilados**

**Gráficos circulares**

```python
# Gráfico circular, donde las porciones se ordenan y trazan en sentido antihorario
etiquetas = 'C', 'Python', 'Java', 'C++', 'C#'
tamaños = [13.38, 11.87, 11.74, 7.81, 4.41]
explotar = (0, 0.1, 0, 0, 0) # solo "explotar" la segunda porción (i.e. 'Python')
fig1, ax1 = plt.subplots()
ax1.pie(tamaños, explode=explotar, labels=etiquetas, autopct='%1.1f%%',
shadow=True, startangle=0)
ax1.axis('equal') # La relación de aspecto uniforme garantiza que se dibuje el gráfico circular.
plt.title('Índice TIOBE para Mayo 2021')
plt.show()
```

```python
# Gráfico circular, donde las porciones se ordenan y trazan en sentido antihorario:
etiquetas = 'C', 'Python', 'Java', 'C++', 'C#', 'otros'
tamaños = [13.38, 11.87, 11.74, 7.81, 4.41]
tamaños.append(100 - sum(tamaños))
explotar = (0, 0.1, 0, 0, 0, 0) # solo "explotar" la segunda porción (i.e. 'Python')
fig1, ax1 = plt.subplots()
ax1.pie(tamaños, explode=explotar, labels=etiquetas, autopct='%1.1f%%',
shadow=True, startangle=0)
ax1.axis('equal') # La relación de aspecto uniforme garantiza que se dibuje el gráfico circular.
plt.title('Índice TIOBE para Mayo 2021')
plt.show()
```
