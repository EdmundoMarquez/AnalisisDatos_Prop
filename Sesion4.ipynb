```python
#Tipos de Datos
nombre_tienda = "Tienda Sara" #String
abierta = True #Booleano
meta_ventas_diaria = 100.0 #float
dias = ("Lun", "Mar", "Mie", "Jue", "Vie") #tuple

ventas_semana = [
    [40, 30, 20],
    [100, 300, 200],
    [250, 180, 200],
    [0, 0, 0],
    [800, 200, 320]
] #Lista de listas

info = {"tienda": nombre_tienda, "meta": meta_ventas_diaria} #diccionario
categorias = {"bebidas", "snacks", "fruta"} #data set

print("Información: ", info)
print("Categorías: ", categorias)
```

```python
#Operadores Aritméticos
total_semana = 0
totales_por_dia = []

for dia in range(len(ventas_semana)):
    total_dia = ventas_semana[dia][0] + ventas_semana[dia][1] + ventas_semana[dia][2]
    total_semana += total_dia
    totales_por_dia.append(total_dia)

print("Total por día", totales_por_dia)
print("Total de la semana", total_semana)
```

```python
#IF ELIF ELSE
for i in range(len(totales_por_dia)):
    if totales_por_dia[i] == 0:
        print(dias[i], "Sin ventas")
    elif totales_por_dia[i] >= meta_ventas_diaria:
        print(dias[i], "Meta alcanzada")
    else:
        print(dias[i], "Meta no alcanzada")

#FOR ANIDADO
max_venta = 0

for fila in ventas_semana:
    for venta in fila:
        if venta > max_venta:
            max_venta = venta
        
print("Venta máxima registrada:", max_venta)
```
